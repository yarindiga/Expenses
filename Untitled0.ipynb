```python
import pandas as pd
import plotly.graph_objects as go

# Create empty lists to store latitudes and longitudes
lats = []
lons = []

# Open the file and read each line
with open('SerialData_2.txt', 'r') as file:
    for line in file:
        # Extract lat and long values from the line
        lat = float(line.split()[3])
        lon = float(line.split()[5])
        
        # Append lat and long values to the lists
        lats.append(lat)
        lons.append(lon)

# Set the coordinates of Rehovot, Israel
rehovot = {'lat': 31.8948, 'lon': 34.8093}

# Create a scattermapbox plot with the latitudes and longitudes
fig = go.Figure()

# Add a trace with mode 'lines+markers' to show the route
fig.add_trace(go.Scattermapbox(
    lat=lats,
    lon=lons,
    mode='lines+markers',
    marker=go.scattermapbox.Marker(
        size=9,
        color='red'
    ),
    line=dict(width=4, color='blue'),
    text=['Lat: ' + str(lat) + '<br>Long: ' + str(lon) for lat, lon in zip(lats, lons)]
))

# Set the map's view to Rehovot, Israel
fig.update_layout(
    mapbox_style="open-street-map",
    mapbox=dict(center=rehovot, zoom=13),
    margin=dict(l=0, r=0, t=0, b=0),
)

fig.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# Create an empty list to store the points
points = []

# Read the file and extract the latitudes and longitudes
with open('SerialData_2.txt') as f:
    for line in f:
        match = re.match(r'Time: .* Lat: (\d+) Long: (\d+)', line)
        if match:
            latitude = int(match.group(1)) / 10**7  # Convert to decimal degrees
            longitude = int(match.group(2)) / 10**7
            point = Point(longitude, latitude)
            points.append(point)

# Create a GeoDataFrame with the points
gdf = gpd.GeoDataFrame(geometry=points)

# Load the Israel shapefile
israel = gpd.read_file('MAP22-polygon.shp')

# Plot the map
ax = israel.plot(color='white', edgecolor='black', figsize=(10, 10))
gdf.plot(ax=ax, color='red', markersize=5)
plt.show()
```
